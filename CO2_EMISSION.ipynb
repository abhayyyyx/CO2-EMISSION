```python
#importing libraries 
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

LINEAR REGRESSION

```python
def train_linear_model(X_train, y_train):
    model = LinearRegression().fit(X_train,y_train)
    return model
# Evualting the regression model
def evaluate_linear(model, X_train,y_train,X_test,y_test):
    train_rsme = np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
    test_rsme = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
    r2_train = model.score(X_train,y_train)
    lin_r2_score = cross_val_score(model, X_train, y_train, cv=10, scoring="r2").mean()
    return train_rsme, test_rsme, r2_train, lin_r2_score

# Predicting and comparing the model
def predict_compare(model, sample_df_x, sample_df_y):
    pred = model.predict(sample_df_x)
    result_pred = pd.DataFrame({"Pred_linear":pred, "real_value":sample_df_y.values})
    result_pred["Pred_linear"] =  result_pred['Pred_linear'].round(2)
    result_pred["diff"] =  result_pred['Pred_linear'] - result_pred['real_value']
    mean_diff = abs(result_pred["diff"]).mean()
    return result_pred, mean_diff


# As we've defined the train and test value, we can check the values of it
if __name__ == "__main__":
    model = train_linear_model(X_train, y_train) 
    train_rsme, test_rsme, r2_train, lin_r2_score = evaluate_linear(model, X_train, y_train, X_test, y_test)
    print("Training Root sqaure mean error with linear regression :",train_rsme)
    print("Test root square mean error with linear regression :",test_rsme)
    print("R^2 score on training data with linear regression :",r2_train)
    print("Cross Validation of R^2 score with linear regression :",lin_r2_score)
    linear_regression_pred, mean_diff = predict_compare(model, sample_df_x, sample_df_y)
    print("The mean difference : ",mean_diff)
```

RANDOM FOREST

```python
def train_random_forest(X_train, y_train):
    model = RandomForestRegressor().fit(X_train,y_train)
    return model

# Evaulating random forest model
def evaluate_random(model, X_train, y_train, X_test, y_test):
    train_rsme = np.sqrt(mean_squared_error(y_train,model.predict(X_train)))
    test_rsme = np.sqrt(mean_squared_error(y_test,model.predict(X_test)))
    r2_score = model.score(X_train,y_train)
    rf_r2_score = cross_val_score(model, X_train, y_train, cv=10, scoring="r2").mean()
    return train_rsme, test_rsme, r2_score, rf_r2_score

# Predicting and comparing the model

def predict_compare(model, sample_df_x, sample_df_y):
    pred = model.predict(sample_df_x)
    result_pred = pd.DataFrame({"RF_pred":pred,"Real Value":sample_df_y.values})
    result_pred["RF_pred"] = result_pred["RF_pred"].round(2)
    result_pred["Diff"] = result_pred["RF_pred"] - result_pred["Real Value"]
    mean_diff = abs(result_pred["Diff"]).mean()
    return result_pred, mean_diff

# As we've defined the train and test value, we can check the values of it

if __name__ == "__main__":
    model = train_random_forest(X_train,y_train)
    train_rsme, test_rsme, r2_score, rf_r2_score = evaluate_random(model, X_train, y_train, X_test, y_test)
    print("Training Root sqaure mean error with Random Forest :",train_rsme)
    print("Test root square mean error with Random Forest :",test_rsme)
    print("R^2 score on training data with Random Forest :",r2_train)
    print("Cross Validation of R^2 score with Random Forest :",rf_r2_score)
    random_forest_pred, mean_diff = predict_compare(model, sample_df_x, sample_df_y)
    print("Mean difference : ",mean_diff)
```

 KNN REGRESSION MODEL

```python
def knn_model(X_train, y_train):
    model = KNeighborsRegressor().fit(X_train, y_train)
    return model

#Evaluarting the knn model

def evaluate_knn(model, X_train, y_train, X_test, y_test):
    train_rsme = np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
    test_rsme = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
    r2_score = model.score(X_train,y_train)
    knn_r2_score = cross_val_score(model, X_train, y_train, cv=10,scoring="r2").mean()
    return train_rsme,test_rsme, r2_score, knn_r2_score

#Comparing and predicting the model

def predict_compare(model, sample_df_x, sample_df_y):
    pred = model.predict(sample_df_x)
    result_pred = pd.DataFrame({"Pred_knn" : pred, "Real Value":sample_df_y.values})
    result_pred["Pred_knn"] = result_pred["Pred_knn"].map(lambda x:round(x,2))
    result_pred["Diff"] = result_pred["Pred_knn"] - result_pred["Real Value"]
    mean_diff = abs(result_pred["Diff"]).mean()
    return result_pred, mean_diff

# As we've defined the train and test value, we can check the values of it
if __name__ == "__main__":
    model = knn_model(X_train, y_train)
    train_rsme, test_rsme, r2_score, knn_r2_score = evaluate_knn(model, X_train, y_train, X_test, y_test)
    print("Training Root sqaure mean error with KKN model :", train_rsme)
    print("Testt root square mean error with KNN model", test_rsme)
    print("R2 score with KNN model :", r2_score)
    print("Cross validation of r2 score with KNN model :", knn_r2_score)
    knn_model_pred , mean_diff = predict_compare(model, sample_df_x, sample_df_y)
    print("THe mean differnece is :",mean_diff)
```
